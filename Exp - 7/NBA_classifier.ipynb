# Naive Bayesian Algorithm

Write a program to implement the naive Bayesian classifier for a sample training data set stored as a .CSV file. Compute the accuracy of the classifier, considering few test data sets. 

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
```

```python
df = pd.read_csv('pima_indian.csv')
df.head()
```

```python
df.shape
```

```python
feature_col_names = ['num_preg','glucose_conc','diastolic_bp','thickness','insulin',
                    'bmi','diab_pred','age']
```

```python
predicted_class_names = ['diabetes']
```

```python
x = df[feature_col_names].values
x
```

```python
y = df[predicted_class_names].values
y
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.20)
```

```python
print("\nThe total number of training Data : ",ytrain.shape)
print("\nThe total number of test data :",ytest.shape)
```

```python
clf = GaussianNB().fit(xtrain,ytrain.ravel())

predicted = clf.predict(xtest)
predicted
```
